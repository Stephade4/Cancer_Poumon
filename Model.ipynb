```python
from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix,log_loss 
```

```python
df.Level= df.Level.replace(df.Level.unique(),[1,2,3])
df
```

```python
cor= df.corr(numeric_only=True)
plt.figure(figsize=(20,10))
sns.heatmap(cor, annot=True, cmap="coolwarm", fmt=".2f")
plt.show()
```

```python
df.columns.tolist()
```

```python
view1=['Alcohol use']
view2=['Balanced Diet']
view3=['Smoking','Dust Allergy']
view4=['Shortness of Breath']
view5=['Chest Pain']
view6=['chronic Lung Disease']
view7=['Air Pollution']
view8=['Coughing of Blood']
view9=['Genetic Risk']
view10=['Passive Smoker']
view11=['OccuPational Hazards']
view12=['Fatigue']
view13=['Obesity']

target="Level"

total= view1 + view2 + view3 + view4 + view5 +view6+view7+view8+view9+view10+view11+view12+view13


#liste d'algorithme 
sv= SVC()
logic= LogisticRegression()
randon= RandomForestClassifier()
tree= DecisionTreeClassifier()

model=[sv,logic,randon,tree]


#separation des donnees d'entrainement et de test
x_train,x_test,y_train,y_test = train_test_split(df[total],df[target], test_size=0.2 , random_state=42)


for ele in model:
    print(f"RAPPORT DE CLASSIFICATION DE L'element: {ele}")
    ele.fit(x_train, y_train)
    y_pred= ele.predict(x_test)
    #y_pred_prob= ele.predict_proba(x_test)
    print("accuracy : ", accuracy_score(y_test,y_pred))
    #print("Log Loss : ", log_loss(y_test,y_pred_proba))
    print("class report : ", classification_report(y_test,y_pred))
```

```python
from sklearn.model_selection import learning_curve

def curve(model,X,y):
    model_name= model
    train_size,train_score,test_score= learning_curve(model,X,y,cv=5, scoring="accuracy", n_jobs=-1, train_sizes =np.linspace(0.1,1.0,10))
    train_mean= np.mean(train_score,axis=1)
    test_mean= np.mean(test_score,axis=1)
    plt.plot(train_size, train_mean, label="training_score", color="blue")
    plt.plot(train_size, test_mean ,label="validation score",color="red")
    plt.xlabel("Training Score")
    plt.ylabel("Accuracy Score")
    plt.title(f"Training curve : {model_name}")
    plt.legend()
    plt.show()
    plt.grid()
```

```python
for ele in model:
    curve(ele,x_train,y_train)
```

```python
best_model= randon
```

amelioration des model avec grid_searchcv

```python
# amelioration de randomforest
param_grid={
    "n_estimators":[50,75,100,150,200],
    "max_depth": [None,5,10,20],
    "min_samples_split": [2,5,7]
}

grid= GridSearchCV(
    estimator= best_model,
    param_grid= param_grid,
    cv=5,
    scoring="accuracy",
    n_jobs=-1
)

grid.fit(x_train,y_train)

print("le meilleur parmetre est :", grid.best_params_)
print("le meilleur score est :", grid.best_score_)
```

```python
best=grid.best_estimator_
y_val=best.predict(x_test)

print("Accuracy Score : ", accuracy_score(y_test,y_val))
print("classification report : ", classification_report(y_test,y_val))
```

```python
curve(best,x_train,y_train)
```
